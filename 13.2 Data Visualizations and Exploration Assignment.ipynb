```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset("tips")
print(tips.head(n=4))
```

### 1a. Choose one variable and plot that variable four different ways.

```python
#style 
sns.set(style="whitegrid")

# 1.scatterplot
plt.plot(tips["tip"],marker='o', linestyle='')
plt.xlabel("Tips")
plt.title('1. Scatterplot - Number of Tips')
plt.show()

# 2.violinplot
b=sns.violinplot(x=tips["tip"], split=True, inner="quart", data=tips)
b.set(xlabel='Tips', ylabel='')
plt.title('2. Violinplot - Number of Tips')
plt.show()

# 3.boxplot
c = sns.boxplot(x=tips["tip"],data=tips)  
plt.title('3. Boxplot - Number of Tips')
sns.despine(offset=10, trim=True)
c.set(xlabel='Tips', ylabel='')
plt.show()

# 4.swarmplot
d = sns.swarmplot(x=tips["tip"], data=tips)
d.set(xlabel='Tips', ylabel='')
plt.title('4. Swarmplot - Number of Tips')
plt.show()

print('')
print('')
```

### 1b. Choose two continuous variables, and plot them three different ways.

```python
#1. Lineplot - Tips and Total Bill
sns.set(color_codes=True)
a = sns.lineplot(x="tip", y="total_bill", palette="dark", data=tips)
a.set(xlabel='Tips', ylabel='Total Bill')
plt.title('1. Lineplot - Tips and Total Bill')
plt.show()

#2. Regplot - Tips and Total Bill 
b = sns.regplot(x="tip", y="total_bill", data=tips)
b.set(xlabel='Tips', ylabel='Total Bill')
plt.title('2. Regplot - Tips and Total Bill')
plt.show()

#3. Scatterplot - Tips and Total Bill
c = sns.relplot(x="tip", y="total_bill", size="size", hue="size", data=tips)
c.set(xlabel='Tips', ylabel='Total Bill')
plt.title('3. Scatterplot - Tips and Total Bill')
plt.show()

print('')
print('')
```

### 1c. Choose one continuous variable and one categorical variable, and plot them six different ways.


```python
#style 
sns.set(style="darkgrid",palette="pastel")


# 1.barplot
a = sns.barplot(x="sex", y="tip", data=tips)
plt.title('1. Barplot - Meal Tips by Gender')
a.set(xlabel='Sex', ylabel='Tips')
plt.show()


# 2.violinplot
b = sns.violinplot(x="sex", y="tip", split=True, inner="quart", data=tips)
b.set(xlabel='Sex', ylabel='Tips')
plt.title('2. Violinplot - Meal Tips by Gender')
plt.show()


# 3.boxplot
c = sns.boxplot(x='sex',y='tip',data=tips)  
plt.title('3. Boxplot - Meal Tips by Gender')
sns.despine(offset=10, trim=True)
c.set(xlabel='Sex', ylabel='Tips')
plt.show()


# 4.swarmplot
d = sns.swarmplot(x="sex", y="tip", data=tips)
d.set(xlabel='Sex', ylabel='Tips')
plt.title('4. Swarmplot - Meal Tips by Gender')
plt.show()


men = tips[(tips['sex'] == "Male")]
women = tips[(tips['sex'] == "Female")]

#5 Scatterplot
plt.plot(men['tip'], '.', color='orange')
plt.plot(women['tip'], '.', color='lightskyblue')
plt.yticks([0,2,4,6,8,10])
plt.title('5. Scatterplot - Meal Tips by Gender')
plt.ylabel('Tip Value')
plt.show()


#6 Lineplot
plt.plot(men['tip'], color='orange')
plt.plot(women['tip'], color='lightskyblue')
plt.yticks([0,2,4,6,8,10])
plt.title('6. Lineplot - Meal Tips by Gender')
plt.ylabel('Tip Value')
plt.show()


#7 Histogram
plt.hist(men['tip'], bins=20, color='orange', alpha=0.6)
plt.hist(women['tip'], bins=20, color='blue',alpha=0.9)
plt.title('#7 Histogram - Meal Tips by Gender')
plt.axvline(men['tip'].mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(men['tip'].mean() + men['tip'].std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(men['tip'].mean()- men['tip'].std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(women['tip'].mean(), color='cyan', linestyle='solid', linewidth=1)
plt.axvline(women['tip'].mean() + women['tip'].std(), color='cyan', linestyle='dashed', linewidth=1)
plt.axvline(women['tip'].mean() - women['tip'].std(), color='cyan', linestyle='dashed', linewidth=1)
plt.xlabel('Tips')
plt.show()

print('')
print('')
```

### 1d. Give the pros and cons of each plot you create. You can use variables from multiple datasets if you like.

#### Scatterplots
Pros:
It shows the relationship between two variables.
It is the best method to show you a non-linear pattern.
The range of data flow, i.e. maximum and minimum value, can be easily determined.
Observation and reading is straightforward.
Plotting the diagram is relatively simple.
Cons:
Data points can over lap.
Data points with decimal points can be hard to distinguish from whole intergers 

#### Violinplot
Pros: 
shows multimodal data
shows normality/skewness unambiguously
Cons:
hides n
not familiar to many readers

#### Boxplot
Pros: 
compact
easy to tell differences between groups
shows normality/skewness
Cons: 
hides multimodal data
sometimes obscures the n number
many variations

#### Swarmplot
Pros:
No overlaps
Easily see outliers
Can give a sense of distribution
Cons:
No overlaps means more space required to plot each data point
not familiar to many readers

#### Lineplot
Pros:
Give quick analysis of data
The range of data flow, i.e. maximum and minimum value, can be easily determined.
changes over a certain period of time
Cons:
Too many data points make it hard to read

#### Regplot
Scatterplot that assumes a linear relationship between the input variables (x) and the single output variable (y)

#### Bargraph (single)
Pros:
Great for displaying different catagories
Easily compare several data sets
straight forward
Cons:
can only be used with discrete data (data that only takes on certain values)
It can also be misleading, since if you organize a bar graph a certain way, it could emphasize a certain effect even though that effect isn't actually that statistically relevant

#### Histogram
Pros:
Most commonly used graphs to display data
Easily compare data
Work well with large ranges of information
Cons:
extremely difficult and practically impossible to extract the exact amount of "input"
Inconvenient when comparing multiple categories

```python
df = pd.read_csv('heart.csv')
print(df.head(n=5))
```

```python
sns.set(style="ticks", color_codes=True)

m = sns.pairplot(df, hue="target", markers=["o", "x"],
                 x_vars=["age", "trestbps","chol", "thalach"],
                 y_vars=["age", "trestbps","chol", "thalach"])
new_labels=["No Heart Disease","Has Heart Disease"]
for t, l in zip(m._legend.texts, new_labels): t.set_text(l)
m.fig.suptitle("Heart Disease")

plt.show()
```
