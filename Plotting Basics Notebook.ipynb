```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('Desktop\Thinkful Directory\heart.csv')
```

```python
plt.subplot(2, 1, 1)
plt.plot(df['age'], df['chol'], '.', color='orange')
plt.title('#1 Influence of Age')
plt.ylabel('Cholestrol')

plt.subplot(2, 1, 2)
plt.plot(df['age'], df['trestbps'], '.', color='red')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')

print("In these two graphs it looks like age has a influence towards a person's cholestrol and resting blood pressure. A quick search online tells me that anything above 240 cholestrol for adults are considered very high and I am surprised that there are more people with 200 or lower cholestrol then I expected. There is also seems to ba a strong decline in the data set after the age of 70.")
```

```python
df.plot(kind='scatter',color='purple', x='trestbps',y= 'chol')
plt.title('#2 Resting Blood Pressure vs Cholestrol')
plt.show()

print("This next graph seems to indicate that there is no strong correlation between cholestrol and resting blood pressure. Looking closely at the graph I notice these concentrated lines of data points such as those with trestbps 120, 130, and 140. I wonder when colloecting the data, there was a bias toward rounding the number. This leads me to think that data sets aren't just messy but also show weird patterns. ")
```

```python
df.plot(kind='scatter',color='black', x='thalach',y= 'chol')
plt.title('#3 Maximum Heart Rate Achieved vs Cholestrol')
plt.show()
print("Looking at this graph there doesn't seem to be a clear indication that cholestrol levels affect a persons maximum heart rate. Though some of the highest cholestrol levels seem to peak at 160 while the highest heart rates are between 200 and 300 cholestrol levels. I wonder if these numbers suggest that in the extremes, cholestrol can limit a persons maximum heart rate.")
```

```python
df.plot(kind='scatter',color='blue', x='age',y= 'thalach')
plt.title('#4 Age and Maximum Heart Rate')
plt.show()

print("Age and maximum heart rate seems have a postive correlation. It seems in this graph that only till the age of 53 can people reach the lowest maximum heart rates. It makes me think that 50 is a much more important age then I would like to believe. ")
```
