```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

## Question 1 - Part 1: Increase the size of your samples from 100 to 1000, then calculate the means and standard deviations for your new samples and create histograms for each. 

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()


plt.figure(figsize=(10, 5)) 

sample1k = np.random.choice(pop1, 1000, replace=True)
sample2k = np.random.choice(pop1, 1000, replace=True)

plt.subplot(1, 2, 1)
plt.hist(sample1k, bins=20, color = 'c')
plt.axvline(sample1k.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(sample1k.mean() + sample1k.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(sample1k.mean() - sample1k.std(), color='b', linestyle='dashed', linewidth=2)

plt.subplot(1, 2, 2)
plt.hist(sample1k, bins=20, color = 'orange')
plt.axvline(sample2k.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(sample2k.mean() + sample2k.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(sample2k.mean() - sample2k.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

print('Sample 1 mean: ' + str(sample1k.mean()))
print('Sample 2 mean: ' + str(sample2k.mean()))
print('Sample 1 mean: ' + str(sample1k.std()))
print('Sample 2 mean: ' + str(sample1k.std()))
print()
```

```python
print('Increasing the sample from 100 to 1000 will better reflect the population. Decreasing the sample size to 20 will have the inverse effect and have a less accurate representation of the population')
```

## Question 1 - Part 2: Repeat this again, decreasing the size of your samples to 20. What values change, and what remain the same?

```python
plt.figure(figsize=(10, 5)) 

sample1_20 = np.random.choice(pop1, 20, replace=True)
sample2_20 = np.random.choice(pop1, 20, replace=True)

plt.subplot(1, 2, 1)
plt.hist(sample1k, bins=20, color = 'c')
plt.axvline(sample1_20.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(sample1_20.mean() + sample1_20.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(sample1_20.mean() - sample1_20.std(), color='b', linestyle='dashed', linewidth=2)

plt.subplot(1, 2, 2)
plt.hist(sample1_20, bins=20, color = 'orange')
plt.axvline(sample2_20.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(sample2_20.mean() + sample2_20.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(sample2_20.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

print('Sample 1 mean: ' + str(sample1_20.mean()))
print('Sample 2 mean: ' + str(sample2_20.mean()))
print('Sample 1 mean: ' + str(sample1_20.std()))
print('Sample 2 mean: ' + str(sample1_20.std()))
print()
```

## Question 2 - Part 1: Change the probability value (p in the NumPy documentation) for pop1 to 0.3, then take new samples and compute the t-statistic and p-value. 

```python
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

## Question 2 - Part 2: Then change the probability value p for group 1 to 0.4, and do it again. What changes, and why?

```python
pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```
